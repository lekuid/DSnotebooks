## Q: Develop an association-based rule on the following dataset using Python.

```python
import csv
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
file = open('groceries.csv')
csvreader = csv.reader(file)
rows = []
for row in csvreader:
    rows.append(str(row).replace("[", "").replace("]", "").replace("'", "").split(','))
rows[0]
```

```python
te = TransactionEncoder()
te_data = te.fit_transform(rows)
df = pd.DataFrame(te_data, columns=te.columns_)
df
```

```python
df1 = apriori(
    df,
    min_support=0.01,
    use_colnames=True
    )
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
```

```python
df1[(df1['length']>=3) & (df1['support']>=0.01)]
```
