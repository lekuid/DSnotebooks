```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import normalize
from sklearn.preprocessing import Binarizer
import warnings
```

```python
input_data = np.array([[3, -1.5, 3, -6.4], [0, 3, -1.3, 4.1], [1, 2.3, -2.9, -4.3]])
df = pd.DataFrame(input_data)
```

```python
df.head()
```

# Mean removal

```python
#Subtracting mean of column from values of each column to center the data around a mean of 0.

#declaring new dataFrame
meanRemoved = pd.DataFrame(columns=df.columns)

#column-wise subtracting means from values
for col in df.columns:
        meanRemoved[col] = df[col].apply(lambda x : x-df[col].mean())

#Resulting dataFrame and Means of the resulting dataFrame rounded of to 10places
fig, ax=plt.subplots(1,2)
sns.distplot(df, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(meanRemoved, ax=ax[1])
ax[1].set_title("Mean removed data")

warnings.filterwarnings("ignore")

print(f'Resulting means: \n{list(round(meanRemoved.mean(), 10))}')
```

# Scale the data

```python
#Scaling the dataFrame to desired range

#Creating a scaler object, defining range from -1 to 1 and fitting the dataFrame
scaler = MinMaxScaler(feature_range=(-1,1))
scaler = scaler.fit_transform(df)

#New dataFrame from the scaled values
scaledDF = pd.DataFrame(scaler, columns=df.columns)
print(f'Resulting dataFrame: \n{scaledDF.head()} \n')

fig, ax=plt.subplots(1,2)
sns.distplot(df, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaledDF, ax=ax[1])
ax[1].set_title("Scaled data")

warnings.filterwarnings("ignore")
```

# Normalize the data

```python
#Normalizing the dataFrame

#Using the normalize function to compress values into 0 to 1 range
normalized = normalize(df)

#New dataFrame from the normalized values
normalizedDF = pd.DataFrame(normalized)
print(f'Resulting dataFrame: \n{normalizedDF.head()} \n')

fig, ax=plt.subplots(1,2)
sns.distplot(df, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalizedDF, ax=ax[1])
ax[1].set_title("Normalized data")

warnings.filterwarnings("ignore")
```

# Binarize the data

```python
#Binarizing the dataFrame

#Creating a binarizer object, setting a threshold at 0
binarize = Binarizer(0)
binarize = binarize.fit_transform(df)

#New dataFrame from the scaled values
binarizedDF = pd.DataFrame(binarize, columns=df.columns)
print(f'Resulting dataFrame: \n{binarizedDF.head()} \n')
```
